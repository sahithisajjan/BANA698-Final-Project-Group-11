## 01 — Data Loading and Initial Inspection

This notebook performs:
- Loading of the raw dataset
- Conversion from Excel to CSV
- Initial structural inspection
- Missing value and duplicate checks

This ensures the dataset is correctly prepared before feature engineering and modeling.

## 1. Import Required Libraries

We begin by importing the necessary Python libraries for data manipulation and file handling.

```python
import pandas as pd
import os

print("Libraries loaded successfully.")
```

## 2. Load Raw Dataset (Excel Version)

The original dataset is stored as an Excel file.  
We load it into a pandas DataFrame for inspection.

This is a one-time step before converting it to CSV format.

```python
excel_path = "../Data/raw/dataset_algorithmic_persuasion_10000.xlsx"
df = pd.read_excel(excel_path)

print("Excel file loaded successfully!")
print("Dataset shape (rows, columns):", df.shape)

df.head()
```

## 3. Convert Excel to CSV (Standardization Step)

CSV format is preferred for:
- Faster loading
- Fewer dependency issues
- Better compatibility with version control

The data itself is not modified during this process.

```python
csv_path = "../Data/raw/dataset_algorithmic_persuasion_10000.csv"
df.to_csv(csv_path, index=False)

print("Excel converted to CSV successfully!")
```

## 4. Load Dataset from CSV (Primary Working File)

From this point forward, we use the CSV version of the dataset.

```python
df = pd.read_csv(csv_path)

print("CSV file loaded successfully!")
print("Dataset shape:", df.shape)

df.head()
```

## 5. Inspect Dataset Structure

We examine:
- Column names
- Data types
- Overall dataset structure

This helps ensure variables align with our research constructs.

```python
print("Column Names:")
print(df.columns.tolist())

print("\nDataset Info:")
df.info()
```

## 6. Missing Value Analysis

We calculate missing values across all variables to identify:
- Columns requiring cleaning
- Potential modeling limitations

```python
missing_values = df.isna().sum().sort_values(ascending=False)

print("Top 20 Columns with Missing Values:")
print(missing_values.head(20))
```

## 7. Duplicate Row Check

Duplicate records can bias modeling results.  
We verify whether identical rows exist in the dataset.

```python
duplicates = df.duplicated().sum()

print("Number of duplicate rows:", duplicates)
```

## 8. Save Baseline Processed Version

To preserve the raw dataset, we save a clean baseline copy.
Future notebooks will build features from this processed file.

```python
processed_path = "../Data/processed/baseline_dataset.csv"
df.to_csv(processed_path, index=False)

print("Baseline dataset saved successfully!")
```
